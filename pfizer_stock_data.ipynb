# Factors to Measure and Data to Collect
Customer Satisfaction and Feedback:
Data to Collect: Survey results, Net Promoter Score (NPS), customer reviews, qualitative feedback.
Why: Direct feedback provides insights into customer loyalty and satisfaction.

# Customer Engagement Metrics:
Data to Collect: Frequency of interactions, number of meetings, email exchanges, project milestones.
Why: High engagement often correlates with long-term retention.

# Service Usage Patterns:
Data to Collect: Types of services used, frequency of service utilization, duration of service contracts.
Why: Customers who use more services or renew contracts are more likely to stay.

# Financial Metrics:
Data to Collect: Revenue generated per customer, payment history, profitability of the customer account.
Why: Financially stable relationships are more likely to be sustained.

# Customer Demographics and Firmographics:
Data to Collect: Industry, company size, location, customer tenure.
Why: Different industries or company sizes may have varying retention patterns.

# Competitive Factors:
Data to Collect: Competitor activity, market trends, customer exposure to alternative firms.
Why: External factors can influence a customerâ€™s decision to stay or leave.

# Project Success and Outcomes:
Data to Collect: Project completion rates, achievement of KPIs, ROI for the customer.
Why: Successful projects build trust and loyalty.

# Customer Support and Responsiveness:
Data to Collect: Response times, resolution rates for issues, customer support interactions.
Why: High-quality support enhances satisfaction and reduces churn.

# Contractual and Legal Factors:
Data to Collect: Contract terms, renewal clauses, penalties for early termination.
Why: Legal obligations can influence retention.

# Customer Behavior Trends:
Data to Collect: Historical data on retention, churn rates, reasons for leaving.
Why: Past behavior is a strong predictor of future behavior.

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Fetch Pfizer (PFE) stock data
pfe = yf.Ticker("PFE")

# Download historical data for the last 5 years
stock_data = pfe.history(period="5y")

# Display the first few rows of the dataset
print(stock_data.head())

# Basic information about the dataset
print("\nDataset Info:")
print(stock_data.info())

# Summary statistics
print("\nSummary Statistics:")
print(stock_data.describe())

# Plot the closing price over time
plt.figure(figsize=(12, 6))
plt.plot(stock_data['Close'], label='Closing Price', color='blue')
plt.title('Pfizer (PFE) Stock Price Over the Last 5 Years')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.grid()
plt.show()

# Calculate and plot daily returns
stock_data['Daily Return'] = stock_data['Close'].pct_change()

plt.figure(figsize=(12, 6))
plt.plot(stock_data['Daily Return'], label='Daily Return', color='green')
plt.title('Pfizer (PFE) Daily Returns Over the Last 5 Years')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.legend()
plt.grid()
plt.show()
```
